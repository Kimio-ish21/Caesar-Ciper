```python
def encoder(message, shift):
  message = message.lower() #Making it all lowercase
  shift = shift%26 #Wrapping around for over 26 shift
  alphabet = "abcdefghijklmnopqrstuvwxyz"
  shift_dict = {}

  for i in range (len(alphabet)): #Each letter is now set to a new letter through a dictionary and wraps around past z
    changeletter = alphabet[(i + shift) %26]
    shift_dict[alphabet[i]] = changeletter
  hiddenmessage = ""

  for letter in message: #Goes through all the letters from the input and puts it to the new letter - I got help from Sanjay with this

    if letter in shift_dict:
     new_char = shift_dict[letter] #Finds original letter and shifts it to new one
     hiddenmessage += new_char #Goes on to next letter
    else:
      hiddenmessage += letter #If it isn't a letter, nothing happens

  print(hiddenmessage)
  return hiddenmessage



#This is basically the same thing but minus for the shift
def decoder(message, shift):
  message = message.lower()
  shift = shift%26
  alphabet = "abcdefghijklmnopqrstuvwxyz"
  shift_dict = {}

  for i in range (len(alphabet)):
    changeletter = alphabet[(i - shift) %26]
    shift_dict[alphabet[i]] = changeletter
  hiddenmessage = ""

  for letter in message:

    if letter in shift_dict:
     new_char = shift_dict[letter]
     hiddenmessage += new_char
    else:
      hiddenmessage += letter

  print(hiddenmessage)
  return hiddenmessage


x = int(input('Press 1 if you would like to encrypt and 2 if you would like to decrypt\n'))
if x == 1:
  message = input("What's your message\n")
  try:
    shift = int(input("How many would you like to shift this\n"))
    result = encoder(message, shift)
  except ValueError: #Make sure its a number and catches error
    print('Enter a number')
    pass #I looked up how to repeat it again
elif x == 2:
  message = input("What's your message\n")
  try:
    shift = int(input("How many would you like to shift this\n"))
    result = decoder(message, shift)
  except ValueError:
    print('Enter a number')
    pass
else:
  print("Try a valid number")
```
